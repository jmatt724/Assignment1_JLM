## Imputing missing values

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy = 'most_frequent')
```

```python
data['Age'] = imputer.fit_transform(data[['Age']])
```

```python
data.info()
```

```python
data = data.drop(['Name', 'Ticket', 'Embarked'], axis = 1)
```

### I removed name, ticket, and embarked because intuitively they do not seem likely to contribute to survival. Ticket is better described with pclass and fare.

## Getting Dummy Variables

```python
data = pd.get_dummies(data[['PassengerId', 'Survived', 'Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Fare']],
      drop_first = True)
```

```python
data.head()
```

## Finding a Correlation

```python
sns.pairplot(data=data)
```

### Correlation Matrix

```python
corr_matrix = data.corr()
print(corr_matrix)
```

### Findings: From the correlation matrix and the pairplots We can see that the independent variables that contribute the most to if a person did or did not surivive was gender, fare, and their pclass.

```python
data['New_Age'] = pd.cut(data['Age'], bins=[0, 13, 26, 39, 52, 65, 80, np.inf],
                     labels = [0, 13, 26, 39, 52, 65, 80])
```

### Splitting the Data into training and test sets

```python
from sklearn.model_selection import train_test_split

strat_training_set, strat_test_set = train_test_split(data, test_size=0.2, stratify=data['New_Age'], random_state = 42)
```

### remove New_Age

```python
for cat in strat_training_set, strat_test_set:
    cat.drop('New_Age', axis = 1, inplace = True)
```
